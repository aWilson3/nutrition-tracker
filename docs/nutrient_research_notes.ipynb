<h3 style="color:green">Linoleic Acid</h3>

- polyunsaturated omega-6 fatty acid
- sometimes denoted as 18:2 (n-6) or 18:2 cis-9,12
- one of two essential fatty acids for humans, who must obtain it through diet
- typically occurs in nature as a triglyceride rather than a free fatty acid

source: https://en.wikipedia.org/wiki/Linoleic_acid

<h3 style="color:green">$\alpha$-Linolenic Acid</h3>

- polyunsaturated omega-3 fatty acid
- sometimes denoted as 18:3 (n-3) or 18:3 cis-9,12,15
    - this is its lipid number
    - 18 is number of carbon atoms and 3 is number of double bonds
    - 9,12,15 are positions of the double bonds in the fatty acid chain
- essential fatty acid
- carboxylic acid
- In addition to their structural role in cell membranes, omega-3s (along with omega-6s) provide energy for the body and are used to form eicosanoids. Eicosanoids are signaling molecules that have similar chemical structures to the fatty acids from which they are derived; they have wide-ranging functions in the body’s cardiovascular, pulmonary, immune, and endocrine systems
- Some researchers propose that the relative intakes of omega-6s and omega-3s—the omega-6/omega-3 ratio—may have important implications for the pathogenesis of many chronic diseases, such as cardiovascular disease and cancer, but the optimal ratio—if any—has not been defined
    - Most agree that raising EPA and DHA blood levels is far more important than lowering linoleic acid or arachidonic acid levels
    - Because both classes of fatty acids [n-3 and n-6] compete for the same desaturation enzymes, ALA is a competitive inhibitor of linoleic acid metabolism and vice versa
- NIH DRIs for omega-3 is Adequate Intake (AI) as insufficient data was available to develop a Recommended Dietary Allowance (RDA).


source: https://en.wikipedia.org/wiki/Alpha-Linolenic_acid<br>
        https://ods.od.nih.gov/factsheets/Omega3FattyAcids-HealthProfessional/

<h3 style="color:green">Vitamin A</h3>

- Group of fat-soluble retinoids
- Involved in immune function, vision, reproduction, and cellular communication
- Two forms of vitamin A are available in the human diet
    - performed vitamin A
        - found in foods from animal sources
    - provitamin A carotenoids
        - found in plant pigments
        - most important of this type 'by far' is beta-carotene
- RDAs for vitamin A are given as retinol activity equivalents (RAE) to account for the different bioactivities of retinol and provitamin A carotenoids, all of which are converted by the body into retinol
- One mcg RAE is equivalent to 1 mcg retinol, 2 mcg supplemental beta-carotene, 12 mcg dietary beta-carotene, or 24 mcg dietary alpha-carotene or beta-cryptoxanthin
- Vitamin A is now measured in mcg RAE, but it was previously measured in International Units (IUs)
- Conversions:
    - 1 IU retinol = 0.3 mcg RAE
    - 1 IU supplemental beta-carotene = 0.3 mcg RAE
    - 1 IU dietary beta-carotene = 0.05 mcg RAE
    - 1 IU dietary alpha-carotene or beta-cryptoxanthin = 0.025 mcg RAE

*Note - Due to the difference conversion factors for different vitamin forms, a straightforward conversion from IU to RAE may be difficult if our data does not give that level of granularity.

<h3 style="color:green">Vitamin D</h3>

- Also referred to as calciferol
- Fat-soluble vitamin naturally present in a few foods
- Added to other foods and available as a dietary supplement
- Also produced endogenously when UV rays from sunlight strike the skin and trigger vitamin D synthesis
- Vitamin D promotes calcium absorption in the gut and maintains adequate serum calcium and phosphate concentrations to enable normal bone mineralization and to prevent hypocalcemic tetany (involuntary contraction of muscles, leading to cramps and spasms)
- It is also needed for bone growth and bone remodeling by osteoblasts and osteoclasts
- Together with calcium, vitamin D also helps protect older adults from osteoporosis.
- Vitamin D has other roles in the body, including reduction of inflammation as well as modulation of such processes as cell growth, neuromuscular and immune function, and glucose metabolism
- In foods and dietary supplements, vitamin D has two main forms, D2 (ergocalciferol) and D3 (cholecalciferol), that differ chemically only in their side-chain structures. Both forms are well absorbed in the small intestine.

<h3 style="color:green">Vitamin E</h3>

- Collective name for a group of fat-soluble compounds with distinctive antioxidant activities
    - Antioxidants protect cells from the damaging effects of free radicals, which are molecules that contain an unshared electron. 
    - Unshared electrons are highly energetic and react rapidly with oxygen to form reactive oxygen species (ROS)
    - Vitamin E is a fat-soluble antioxidant that stops the production of ROS formed when fat undergoes oxidation. Scientists are investigating whether, by limiting free-radical production and possibly through other mechanisms, vitamin E might help prevent or delay the chronic diseases associated with free radicals
- In addition to its activities as an antioxidant, vitamin E is involved in immune function and, as shown primarily by in vitro studies of cells, cell signaling, regulation of gene expression, and other metabolic processes
- 
- Found naturally in some foods, added to others, and available as a dietary supplement.
- Naturally occuring Vitamin E exists in eight chemical forms
    - only alpha-tocopherol is recognized to meet human requirements
- Acknowledging “great uncertainties” in these data [used to develop recommendations], the FNB has called for research to identify other biomarkers for assessing vitamin E requirements.

source: https://ods.od.nih.gov/factsheets/VitaminE-HealthProfessional/

<h3 style="color:green">The Effects of Cooking</h3>

- Because vitamin C is water-soluble and sensitive to heat, it can leach out of vegetables when they’re immersed in hot water.
- B vitamins are similarly heat sensitive. Up to 60% of thiamine, niacin, and other B vitamins may be lost when meat is simmered and its juices run off.
- However, when the liquid containing these juices is consumed, 100% of the minerals and 70–90% of B vitamins are retained
- up to 40% of B vitamins and minerals may be lost during grilling or broiling when the nutrient-rich juice drips from the meat
- Short cooking times and reduced exposure to heat preserve the nutrients in microwaved food 
- Most vitamin losses are minimal with roasting/baking, including vitamin C.
    - However, due to long cooking times at high temperatures, the B vitamins in roasted meat may decline by as much as 40%
- In general, sauteing/stir frying is a healthy way to prepare food.
    - Cooking for a short time without water prevents the loss of B vitamins, and the addition of fat improves the absorption of plant compounds and antioxidants
- Fatty fish are the best sources of omega-3 fatty acids, which have many health benefits. However, these fats are very delicate and prone to damage at high temperatures.
    - In contrast, frying preserves vitamin C and B vitamins, and it may also increase the amount of fiber in potatoes by converting their starch into resistant starch
- Steaming is one of the best cooking methods for preserving nutrients, including water-soluble vitamins, which are sensitive to heat and water
    - Researchers have found that steaming broccoli, spinach, and lettuce reduces their vitamin C content by only 9–15%
<br><br>
- The process of cooking food breaks down some of its fibers and plant cell walls, making it easier for the body to digest and absorb the nutrients
- Properly cooking grains and legumes not only improves their digestibility, but it also reduces the number of anti-nutrients they contain. Anti-nutrients are compounds that inhibit the body’s ability to absorb nutrients in plant foods
<br><br>
- Raw or undercooked legumes contain dangerous toxins called lectins. Lectins are eliminated with proper soaking and cooking.
- Cooking mushrooms helps degrade agaritine, a potential carcinogen found in mushrooms. Cooking also helps release ergothioneine, a powerful mushroom antioxidant 
- Other foods more nutritious when cooked:
    - asparagus
    - spinach
    - tomatoes
    - carrots
- Foods healthier when raw:
    - broccoli
        - Raw broccoli contains three times the amount of sulforaphane, a cancer-fighting plant compound, than cooked broccoli does
    - cabbage
        -  Cooking cabbage destroys the enzyme myrosinase, which plays a role in cancer prevention. If you choose to cook cabbage, do so for short periods
    - onions
    - garlic
        - Sulfur compounds found in raw garlic have anti-cancer properties. Cooking garlic destroys these sulfur compounds 

source: https://www.healthline.com/nutrition/cooking-nutrient-content#sauting-and-stirfrying
